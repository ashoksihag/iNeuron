### 9. How do you make a function variable refer to the global variable?

```python

#use of global keyword
g = 'It is a global variable'

def variable():
    global g
    g = 'It is a modiedied global variable'
    print(g)
    
```

```python
print(g)
variable()
```

```python
 #without use of global keyword redefined as local
    
g = 'It is a global variable'

def variable():
    
    g = 'It is a local variable'
    print(g)
```

```python
print(g)
variable()
```

```python
#directly calling global variable

g = 'It is a global variable'

def variable():
    
    print(g)
    
variable()
```
